```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv("matches.csv")
print(df)
```

```python
# Display the first few rows
df.head()
```

```python
#  Total Runs by Team
total_runs=df.groupby('Team')['Runs'].sum().sort_values(ascending=False)
print("Total runss by team:")
print(total_runs)
```

```python
# After feb
after_feb=df[df['Date']> '2026-02-10']
print("\n Matches after Feb 10, 2026 :")
print(after_feb)
```

```python
#Pie Chart – Top 5 Run Scoring Teams
top5_teams=total_runs.head(5)
print(top5_teams)
plt.Figure(figsize=(9,7))
plt.pie(top5_teams, labels=top5_teams.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Run Scoring Teams")
plt.axis()
plt.show()
```

```python
#Line Chart – Runs Over Time for Team India
# Filter only India's matches
india_df = df[df['Team'] == 'India'].copy()
india_df['Date']=pd.to_datetime(india_df['Date'])
# Sort by date (important for line plot)
india_df=india_df.sort_values('Date')
```

```python
# Plot

plt.figure(figsize=[10,5])
plt.plot(india_df['Date'], india_df['Runs'], marker='o', linestyle='-', color='blue')
plt.title("India Runs Over Time")
plt.xlabel('Date')
plt.ylabel('Runs')
plt.grid(True)
plt.tight_layout()
plt.show()
```
