```python
#Cricket Match Analysis: Team Performance Dashboard
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('matches.csv')
print(df)
```

```python
# Display the first few rows
df.head()
```

```python
#  Total Runs by Team
total_runs=df.groupby('Team')['Runs'].sum().sort_values(ascending=False)
print("\nTotal Runs by team:\n")
print(total_runs)
```

```python
# After feb
after_feb=df[df['Date']>'2025-02-01']
print("Matches after Feb 1, 2025:")
print(after_feb)
```

```python
#Pie Chart – Top 5 Run Scoring Teams
top5_teams = total_runs.head(5)
print(top5_teams)
plt.figure(figsize=(8, 6))
plt.pie(top5_teams, labels=top5_teams.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Run Scoring Teams')
plt.axis('equal')
plt.show()
```

```python
#Line Chart – Runs Over Time for Team India

team_df=df[df['Team']=='india'].sort_values('Date')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(team_df['Date'], team_df['Runs'], marker='o', linestyle='-', color='blue')
plt.title('India Runs Over Time')
plt.xlabel('Date')
plt.ylabel('Runs')
plt.grid(True)
plt.tight_layout()
plt.show()
```
